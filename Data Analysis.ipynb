```python
a = np.mean(data_0.iloc[:,:].values, axis=0)-np.mean(data_1.iloc[:,:].values, axis=0)
b= np.std(data_0.iloc[:,:].values, axis=0)
c = np.std(data_1.iloc[:,:].values, axis=0)

names = data.columns.tolist()

s = []

for i in range(len(a)):
    s.append((names[i], a[i],b[i],c[i]))
    
sorted(s, key=lambda x: x[1], reverse=True)
```

```python
for j in range(len(s)):
    
    i = s[j]
    
    index = names.index(i[0])
    
    plt.figure(figsize=(8,8))

    h1 = data_0.iloc[:,index]
    h1mean = np.mean(h1)
    h1std = i[2]
    h1 = sorted(list(h1))

    h2 = data_1.iloc[:,index]
    h2mean = np.mean(h2)
    h2std = i[3]
    h2 = sorted(list(h2))
    
    print(i[0])
    print("\t", "mean-diff:", round(i[1],3), "class0-std:", round(i[2],3), "class1-std:", round(i[3],3))
    
    if j < 5:
        pdf = stats.norm.pdf(h1, h1mean, h1std)
        plt.plot(h1, pdf, color='blue', label='class 0')
        pdf2 = stats.norm.pdf(h2, h2mean, h2std)
        plt.plot(h2, pdf2, color='red', label='class 1')
        plt.xlabel("Normalised " + i[0])
        plt.ylabel("Qty")
        
    else:
        plt.hist(h1, bins=[0,1,2] ,color='blue', label='class 0', histtype='step')
        plt.hist(h2, bins=[0,1,2], color='red', label='class 1', histtype='step')
        plt.xlabel(i[0])
        plt.ylabel("Qty")
        
    plt.legend()
    plt.show()
```

```python
# 12 attributes

def is_similar(x1, x2, threshold):
    if x1[12] != x2[12] or x1[5] != x2[5] or x1[6] != x2[6] or x1[7] != x2[7] or x1[8] != x2[8] or x1[9] != x2[9] or x1[10] != x2[10] or x1[11] != x2[11] :
        return False
    if abs(x1[0]-x2[0])> threshold or abs(x1[1]-x2[1])> threshold or abs(x1[2]-x2[2])> threshold or abs(x1[3]-x2[3])> threshold or abs(x1[4]-x2[4])> threshold:
        return False
    return True

def contrast(t, threshold):
    class_0 = t[0]
    class_1 = t[1]
    
    clusters = []
    not_clustered = []
    b_clustered = []
    
    for a in class_0:
        a_in_cluster = False
        
        for b in class_1:
            if is_similar(a,b,threshold):
                exist = False
                for cluster in clusters:
                    a_in = False
                    for c in cluster[0]:
                        if (a==c).all():
                            a_in = True
                            break
                            
                    b_in = False
                    for c in cluster[1]:
                        if (b == c).all():
                            b_in = True
                            break
                    
                    if a_in and b_in:
                        exist = True
                        break
                    elif a_in and not b_in:
                        exist = True
                        cluster[1].append(b)
                        b_clustered.append(b)
                        
                    elif b_in and not a_in:
                        exist = True
                        a_in_cluster = True
                        cluster[0].append(a)
                    
                if not exist:
                    a_in_cluster = True
                    b_clustered.append(b)
                    clusters.append([[a],[b]])
            
        if not a_in_cluster:
            not_clustered.append(a)
            
    for b in class_1:
        clustered = False
        for b2 in b_clustered:
            if (b==b2).all():
                clustered = True
                break
        if not clustered:
            not_clustered.append(b)
            
    return clusters, not_clustered
```
