```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = pd.read_csv('../t1.csv', delimiter = ',').values
x = data[:,:2]
y = data[:,:-1].astype(int)

train = np.random.choice([True, False], len(data), replace=True, p=[0.9, 0.1])
x_train = data[train,:2]
y_train = data[train,-1].astype(int)
x_test = data[~train,:2]
y_test = data[~train,-1].astype(int)
```

```python
logit = linear_model.LogisticRegression(solver = 'liblinear')
logit.fit(x_train, y_train)
```

```python
print(logit.score(x_test,y_test))
pred_y = logit.predict(x_train)
accuracy_score(pred_y, y_train)
```

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python


data = pd.read_csv("../t1.csv", delimiter = ',').values
x = data[:,:-1]
y = data[:,-1]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)

logit = LogisticRegressionCV(Cs =[0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30], solver='liblinear', cv=10)
```

```python
logit.fit(x_train, y_train)
```

```python
logit.score(x_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

logit = linear_model.LogisticRegression(solver="liblinear")

clf = GridSearchCV(logit, 
           {"C": [0.01,0.03,0.1,0.3,1,3,10,30]}, 
           scoring=None, 
           fit_params=None, 
           n_jobs=None, 
           iid='warn', 
           refit=True, 
           cv=10,
           verbose=0, 
           pre_dispatch='2*n_jobs', 
           error_score='raise-deprecating', 
           return_train_score='warn')
```

```python
clf.fit(x_train, y_train)
```

```python
for i in sorted(clf.cv_results_.keys()):
    print(i)
    print('\t',end="")
    print(clf.cv_results_[i])
```
